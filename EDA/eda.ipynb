```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("diabetes_dataset.csv")
```

```python
target_col="Age"
```

```python
if target_col in data.columns:
    sb.pairplot(data,hue=target_col,diag_kind="hist")
else:
    sb.pairplot(data,diag_kind="hist")
plt.show()
```

```python
numeric_cols=data.select_dtypes(include="number").columns
if target_col in data.columns:
    for col in numeric_cols:
        sb.boxplot(x=target_col, y=col, data=data)
        plt.title(f"Boxplot: {col}")
        plt.show()
```

```python
if target_col in data.columns:
    for col in numeric_cols:
        sb.violinplot(x=target_col, y=col, data=data)
        plt.title(f"Violin Plot: {col}")
        plt.show()
```

```python
sb.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
